```python
import pandas as pd
import psycopg2
from sqlalchemy import create_engine

# --------------------------------------------------------
# 1. Leer CSV
# --------------------------------------------------------
csv_path = "/Users/ndev/Desktop/ndev25/universos/SP/spy.csv"
df = pd.read_csv(csv_path)

# Mostrar primeros registros para verificar
df.head()
```

```python
print(df.columns)
```

```python
# --------------------------------------------------------
# 2. Conexión a PostgreSQL
# --------------------------------------------------------
# Ajusta estos parámetros según tu entorno
db_user = "ndev"
db_pass = "ndev"
db_host = "localhost"
db_port = "5433"
db_name = "ndev25"

engine = create_engine(f"postgresql://{db_user}:{db_pass}@{db_host}:{db_port}/{db_name}")

# --------------------------------------------------------
# 3. Insertar tickers en universos.sp500
# --------------------------------------------------------

# Nos quedamos solo con la columna ticker
sp500_tickers = df["ticker"].dropna().unique()

insert_df = pd.DataFrame({"ticker": sp500_tickers})

# Insertar en la tabla
insert_df.to_sql(
    "sp500",
    engine,
    schema="universos",
    if_exists="append",   # agrega
    index=False
)

print("Tickers insertados correctamente en universos.sp500.")
```
