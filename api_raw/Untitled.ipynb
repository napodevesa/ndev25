```python
from dotenv import load_dotenv
import requests
import os

load_dotenv("/Users/ndev/Desktop/ndev25/.env")
API_KEY = os.getenv("FMP_API_KEY")
ticker  = "ACAD"

# ratios-ttm
url = f"https://financialmodelingprep.com/stable/ratios-ttm?symbol={ticker}&apikey={API_KEY}"
r = requests.get(url)
data = r.json()
print("TIPO:", type(data))
if isinstance(data, list):
    print("CAMPOS:", list(data[0].keys()))
else:
    print("CONTENIDO:", data)
```

```python
# key-metrics
url2 = f"https://financialmodelingprep.com/stable/key-metrics?symbol={ticker}&apikey={API_KEY}"
r2 = requests.get(url2)
data2 = r2.json()
print("TIPO:", type(data2))
if isinstance(data2, list):
    print("CAMPOS:", list(data2[0].keys()))
else:
    print("CONTENIDO:", data2)
```
